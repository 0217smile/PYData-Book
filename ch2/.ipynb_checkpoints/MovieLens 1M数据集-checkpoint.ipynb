```python
import pandas as pd
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('movielens/users.dat', sep='::', header=None, names=unames)

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('movielens/ratings.dat', sep='::', header=None, names=rnames)

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('movielens/movies.dat', sep='::', header=None, names=mnames)
movies[:5]
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
data[:5]
```

```python
mean_ratings = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
mean_ratings[:5]
```

```python
ratings_by_title = data.groupby('title').size()
ratings_by_title[:10]
```

```python
active_titles = ratings_by_title.index[ratings_by_title > 250]
active_titles
```

```python
mean_ratings = mean_ratings.ix[active_titles]
mean_ratings
```

```python
top_female_ratings = mean_ratings.sort_values(by='F', ascending=False)
top_female_ratings[:10]
```

```python
#计算男女评分分歧
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
```

```python
sorted_by_diff = mean_ratings.sort_values(by='diff')
sorted_by_diff[:15]
```

```python
sorted_by_diff[::-1][:15]
```

```python
# 根据电影名称分组的得分数据的标准差
rating_std_by_title = data.groupby('title')['rating'].std()
```

```python
# 根据active_titles进行过滤
rating_std_by_title = rating_std_by_title.ix[active_titles]
```

```python
# 根据值对Series进行降序排列
rating_std_by_title.sort_values(ascending=False)[:10]
```
